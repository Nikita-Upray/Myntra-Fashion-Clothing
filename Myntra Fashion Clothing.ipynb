```python
# Renaming

df.rename(columns={"OriginalPrice (in Rs)": "OriginalPrice"}, inplace=True)
```

```python
# A function which calculate discount

def PriceAfterDiscount(DiscountOffer, original_price):
    if "%" in str(DiscountOffer):
        try:
            dis_per = float(DiscountOffer.split('%')[0])
            after_dis_price = original_price - (dis_per / 100) * original_price
            return after_dis_price
        except ValueError:
            return None  # Handle invalid percentage value
    elif str(DiscountOffer) == '0':
        return original_price
    else:
        try:
            dis_amount = float(DiscountOffer.split('Rs. ')[1].replace(' OFF', ''))
            after_dis_price = original_price - dis_amount
            return after_dis_price
        except ValueError:
            return None
    return None  # Handle other cases
```

```python
# Applying the function

df['PriceToPay'] = df.apply(lambda row: PriceAfterDiscount(row['DiscountOffer'], row['OriginalPrice']), axis=1)
```

```python
df['PriceToPay'].to_frame()
```

```python
df['Savings'] = df['OriginalPrice'] - df['PriceToPay']
```

```python
df['Savings'].to_frame()
```

# Insights: Data Visualization

Top 15 Brands based on product count

```python
# Top 15 brands based on the count of products

top_brands = data['BrandName'].value_counts().head(15)

plt.figure(figsize=(12, 6))
sns.barplot(x = top_brands.values, 
            y = top_brands.index, palette = 'viridis')
plt.title('Top 15 Brands by Number of Products')
```

Top 15 Brands with Maximum Original Price

```python
brand_max_price = data.groupby('BrandName')['OriginalPrice'].max().sort_values(ascending=False).head(15)

plt.figure(figsize = (12, 8))
sns.barplot(x = brand_max_price.values, y = brand_max_price.index, palette = 'viridis')

plt.title('Top 15 Brands with Maximum Original Price')
plt.xlabel('Brand Name')
plt.ylabel('Maximum Original Price')
```

Top 15 Brands with Minimum Original Price

```python
brand_min_price = data.groupby('BrandName')['OriginalPrice'].max().sort_values(ascending=True).head(15)

plt.figure(figsize = (12, 8))
sns.barplot(x = brand_min_price.values, y = brand_min_price.index)

plt.title('Top 15 Brands with Minimum Original Price')
plt.xlabel('Brand Name')
plt.ylabel('Minimum Original Price')
```

Gender Preference for different categories

```python
# Group by category and gender, and count the number of products in each group
gender_count = data.groupby(['Category', 'category_by_Gender'])['Product_id'].count().reset_index()
gender_count = gender_count.rename(columns = {'Product_id': 'Count'})

# Pivot the data to get separate columns for male and female counts
gender_count = gender_count.pivot(index = 'Category', columns = 'category_by_Gender', values = 'Count')

# Plot the bar chart
sns.set_style('whitegrid')
fig, ax = plt.subplots(figsize = (10, 6))
gender_count.plot(kind='bar', ax = ax)

ax.set_title('Gender preference for different categories')
ax.set_xlabel('Category')
ax.set_ylabel('Number of products')
plt.xticks(rotation = 45)
```
